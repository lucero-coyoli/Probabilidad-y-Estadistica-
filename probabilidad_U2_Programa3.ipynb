```python
import math as mt
from itertools import permutations

# Definimos los asistentes
asistentes = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j"]

# Definimos n y r
n = len(asistentes)  # Total de asistentes
r = 4  # Número de preguntas, se selecciona un asistente por pregunta

# Cálculo de permutaciones sin repetición
Pn = mt.factorial(n) / mt.factorial(n - r)

print(f"Existen {Pn:.0f} posibles permutaciones del total de {n} asistentes en grupos de {r}.")

# Generamos e imprimimos las permutaciones
permutaciones_asistentes = list(permutations(asistentes, r))

print("\nAquí las permutaciones:")
for permutacion in permutaciones_asistentes:
    print(permutacion)
```
