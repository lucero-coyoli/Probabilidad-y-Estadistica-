## Pràctica 4: Tablas de Frecuencias

```python
import random
#Lista vacìa
edades=[]
#ciclo de 100 numeros aleatorios que respresentan edades
for i in range (100):
#genera un numero aleatorio y guarda en la variable numero_aleatorio
  numero_aleatorio=random.randint(18,70)
#agrega los numeros aleatorios a la lista vacia de edades
  edades.append(numero_aleatorio)
print(edades)
```

```python
import pandas as pd
```

```python
#crear un DataFrame(datos en columnas y rengoles) de pandas con los numeros aleatorios
df= pd.DataFrame(edades,columns=["edad"])
df.head()
```

```python
# Calcular tabla de distribución de frecuencias
tabla_frecuencias = df["edad"].value_counts().reset_index() # Change "Edad" to "edad"
tabla_frecuencias.columns = ["Edad", "Frecuencia"]
tabla_frecuencias = tabla_frecuencias.sort_values(by="Edad")

# Imprimir la tabla de distribución de frecuencias
print("Tabla de Distribución de Frecuencias:")
print(tabla_frecuencias)

```

```python
tabla_frecuencias [ "Frec Relativa" ]= tabla_frecuencias["Frecuencia"]/100
tabla_frecuencias
```

```python
tabla_frecuencias["frec Porcentual"]=tabla_frecuencias["Frecuencia"]/100 *100
tabla_frecuencias
```

```python
tabla_frecuencias["Frec Acumulada"]= tabla_frecuencias["Frecuencia"].cumsum()
tabla_frecuencias
```

## Regla de Struges

c=1+log2(N)

```python
import math
```

```python
print (edades)
```

```python
#crear un Dataframe con las edades
df= pd.DataFrame({"edad":edades})
df
```

```python
#Calcular el numero de intervalos utilizando la regla de Sturges
n= len(edades)
K= int (1 + math.log2(n))
print("valor de n = ", n ,"\nNùmero de intervalos=", K)
```

```python
#crear la tabla de frecuencia
tabla_frecuencia= pd.cut( df[ "edad" ], bins=K).value_counts().reset_index()
tabla_frecuencia.columns=["clase", "Frec"]
tabla_frecuencia
```

```python
#Ordenar la tabla de frecuencia por el intervalo
tabla_frecuencia = tabla_frecuencia.sort_values(by="clase")
tabla_frecuencia
```

```python
# Generar columnas siguientes

tabla_frecuencia ['Frec Rel'] = tabla_frecuencia['Frec'] / n

tabla_frecuencia['Frec Porc'] = tabla_frecuencia ['Frec'] / n * 100

tabla_frecuencia ['Frec Ac'] = tabla_frecuencia['Frec'].cumsum()

tabla_frecuencia['Frec Rel Ac'] = tabla_frecuencia ['Frec Rel'].cumsum()

tabla_frecuencia ['Frec Porc Ac'] = tabla_frecuencia ['Frec Porc'].cumsum()

# Mostrar la tabla de frecuencia

print(tabla_frecuencia)
```

## Pràctica 5: Diagramas de Cajas y Bigotes

```python
import random
edades2 = []

#Hacemos un ciclo para generar 100 numeros aleatorios qe representan la edad de las personas
for i in range(100):

#generamos un numero aleatorio con la funcion randint y lo guardamos en la variable

#numero_aleatorio

    numero_aleatorio = random.randint(20, 87)

#Agregamos el numero aleatorio a la lista de valores

    edades2.append(numero_aleatorio)

print(edades2)
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))

bplot1=ax1.boxplot (edades,
                    vert=True, # vertical box alignment
                    patch_artist=True) # fill with color

ax1.set_title('Edades Encuesta 1')

bplot2= ax2.boxplot (edades2,
                    vert=True, # vertical box alignment
                    patch_artist=True) # fill with color

ax2.set_title('Edades Encuesta 2')
plt.show()
```
