```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr
import statsmodels.api as sm

# ====== Grupo 1 ======
df = pd.read_csv('grupo1.csv')

# Verificar columnas
print(f'Columnas: {df.columns.tolist()}')

# Histograma de Calificaciones
plt.figure(figsize=(8,5))
sns.histplot(df['Calificacion'], kde=True, color='skyblue')
plt.title('Histograma de Calificaciones - Grupo 1')
plt.xlabel('Calificación')
plt.ylabel('Frecuencia')
plt.show()

# Variables
X = df[['HorasEstudio']]
y = df['Calificacion']

# Modelo de Regresión Lineal
model = LinearRegression()
model.fit(X, y)

pendiente = model.coef_[0]
intercepto = model.intercept_
print(f'Ecuación de la recta: y = {pendiente:.2f}x + {intercepto:.2f}')

# Coeficiente de correlación de Pearson
r, p_valor = pearsonr(df['HorasEstudio'], df['Calificacion'])
print(f'Coeficiente de correlación de Pearson: r = {r:.2f}, p-valor = {p_valor:.4f}')

# Gráfico de dispersión con línea de regresión
plt.figure(figsize=(8,5))
sns.scatterplot(x='HorasEstudio', y='Calificacion', data=df, color='navy')
sns.lineplot(x=df['HorasEstudio'], y=model.predict(X), color='red')
plt.title('Regresión Lineal - Grupo 1')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificación')
plt.show()

# Resumen estadístico con statsmodels
X_sm = sm.add_constant(X)
modelo_sm = sm.OLS(y, X_sm).fit()
print(modelo_sm.summary())
```

Grupo 1
1. Interpretación del coeficiente de regresión:
La pendiente es -0.11, lo cual indica que por cada hora adicional de estudio, la calificación disminuye en promedio 0.11 puntos. Esto no tiene sentido lógico esperado y sugiere que no hay una relación real entre estudio y calificación en este grupo.

2. Interpretación del coeficiente de correlación:
El coeficiente de correlación de r = -0.057 indica una relación muy débil y negativa entre horas de estudio y calificación. En otras palabras, no hay relación significativa.

3. ¿El modelo es estadísticamente significativo?
No. El p-valor del modelo es 0.694, mucho mayor que 0.05. Por lo tanto, la relación no es significativa estadísticamente.

4. ¿Qué tan confiable es el modelo?
Muy poco confiable. El valor de R² = 0.003, lo cual indica que el modelo explica apenas el 0.3% de la variabilidad de las calificaciones.

Conclusión:
En el Grupo 1, no se observa ninguna relación real o significativa entre las horas de estudio y las calificaciones. Es probable que otros factores estén afectando el desempeño.
