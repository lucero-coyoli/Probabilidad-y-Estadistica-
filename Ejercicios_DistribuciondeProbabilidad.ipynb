```python
import matplotlib.pyplot as plt

# Datos para clima soleado
momentos_soleados = [1, 2, 3]
frecuencia_clima = [0.4, 0.2, 0.4]

# Graficar clima soleado
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.bar(momentos_soleados, frecuencia_clima, color='skyblue')
plt.xlabel("Momentos soleados en un día")
plt.ylabel("Frecuencia relativa")
plt.title("Frecuencia relativa del clima")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos para plantas sobrevivientes
plantas_sobrevivientes = [2, 3]
frecuencia_plantas = [0.8, 0.2]

# Graficar plantas sobrevivientes
plt.subplot(1, 2, 2)
plt.bar(plantas_sobrevivientes, frecuencia_plantas, color='lightgreen')
plt.xlabel("Plantas sobrevivientes")
plt.ylabel("Frecuencia relativa")
plt.title("Frecuencia relativa de plantas sobrevivientes")

plt.tight_layout()
plt.show()
```
